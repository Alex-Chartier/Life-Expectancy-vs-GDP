An Exploratory analysis of the life expectancy and GDP of different countries

This project will see the correlation between

    a countries life expectancy trends over the previous 15 years
    a countries GDP trends over the past 15 years
    if there is a correlation between an increase in comparative GDP and an increase in life expectancy
    whether drops in GDP also correlate a reduction in expected life expectancy for their citizens in following years
    If there are common ranges for GDP
    if there are common ranges for Life Expectancy
    if their are common ranges between the two do they line up with each other

GDP Source: World Bank national accounts data, and OECD National Accounts data files.

Life expectancy Data Source: World Health Organization

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import scipy.stats as stats

data = pd.read_csv('all_data.csv')
print(data)
print(data.GDP.mean())
```

```python
data = data.rename({"Life expectancy at birth (years)":"LEABY"}, axis = "columns")
print(data.head())
#Put the value of GDP in millions to make iteasier to grasp the value at a glance
data.GDP = data.GDP.div(1000000)
```

```python
sns.histplot(data = data, x = data.Year, y = data.LEABY, hue = data.Country)
plt.show()
plt.clf()
```

```python
countries = data.Country.unique()
for i in countries:
    print( i + " mean " + str(data.LEABY[data.Country == i].mean()))
    print( i + " max " + str(data.LEABY[data.Country == i].max()))
    print( i + " min " + str(data.LEABY[data.Country == i].min()))
```

```python
sns.histplot(data = data, x = data.Year, y = data.GDP, hue = data.Country, alpha = 0.5)
plt.show()
plt.clf()
```

```python
sns.histplot(data = data, x = data.Year, y = data.GDP, hue = data.Country[data.Country != 'United States of America'])
plt.show()
plt.clf()
```

```python
for i in countries:
    print( i + " mean " + str(data.GDP[data.Country == i].mean()))
    print( i + " max " + str(data.GDP[data.Country == i].max()))
    print( i + " min " + str(data.GDP[data.Country == i].min()))
```

```python
sns.boxplot(data = data, x = data.Country, y = data.GDP)
plt.show()
plt.clf()
```

```python
sns.boxplot(data = data, x = data.Country, y = data.LEABY)
plt.show()
plt.clf()
```
